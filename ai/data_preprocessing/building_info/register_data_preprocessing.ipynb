```python
# `pnu_code`의 각 값에 대해 카운트를 계산
counts = df['pnu_code'].value_counts()

# 한 번만 등장하는 `pnu_code` 값을 가진 행을 필터링
unique_pnu_codes = counts[counts == 1].index

# 유일한 `pnu_code` 값을 가진 행만 선택
unique_rows = df[df['pnu_code'].isin(unique_pnu_codes)].copy()
unique_rows.loc[:, 'main_bld_cnt'] = 1

result_df = unique_rows[['pnu_code','대지면적(㎡)','건축면적(㎡)','건폐율(%)','연면적(㎡)','용적률(%)','주용도코드명','세대수(세대)',
                        'tot_pkng_cnt','대지위치','시군구코드','법정동코드','도로명대지위치',
                       '건물명','생성일자','대장종류코드','대장종류코드명','main_bld_cnt']].copy()

# 날짜 열을 datetime 형식으로 변환
result_df['생성일자'] = pd.to_datetime(result_df['생성일자'], format='%Y%m%d')

# 데이터프레임 컬럼명 영문으로 변경
result_df.rename(columns=field_mapping, inplace=True)

result_df
```

```python
# 'address' 열에서 '번지' 문자열 제거
master_df['대지위치'] = master_df['대지위치'].str.replace('번지', '', regex=False)

# pnu 코드 생성 전에 NaN 값이 있는 행을 제거
master_df.dropna(subset=['시군구코드', '법정동코드', '대지구분코드', '번', '지','주용도코드명'], inplace=True)

# 새로운 pnu 코드 컬럼 추가
master_df['pnu_code'] = master_df.apply(create_pnu, axis=1)

# 일반, 산이 아닌경우 None을 넣어 제거
master_df.dropna(subset=['pnu_code'], inplace=True)

# 데이터프레임 제거 컬럼 선정 후 제거
master_df = master_df[['pnu_code','대지면적', '건축면적', '건폐율', '연면적', '용적률', '주용도코드명', '생성일자', '대장종류코드', '대장종류코드명', '세대수', '주건축물수', '총주차수', '대지위치',
         '시군구코드', '법정동코드', '도로명대지위치', '건물명','가구수','호수','옥내기계식대수','옥외기계식대수','옥내자주식대수','옥외자주식대수']]

# 문자열을 포함하고 있는 열만 확인하여 앞뒤 공백 찍힌거 제거
for column in master_df.columns:
    if master_df[column].dtype == 'object':
        master_df[column] = master_df[column].str.strip()

# 날짜 열을 datetime 형식으로 변환
master_df['생성일자'] = pd.to_datetime(master_df['생성일자'], format='%Y%m%d')

# 각 pnu_code 그룹에서 생성일자의 최대값을 찾기
max_dates = master_df.groupby('pnu_code')['생성일자'].transform('max')

# 최대값과 일치하는 모든 행을 선택하기
master_df = master_df[master_df['생성일자'] == max_dates]

# 생성일자가 같아 2개이상이 반영된 pnu_code를 위해 중복 제거
master_df = master_df.drop_duplicates(subset=['pnu_code'], keep='first')

# 세대수가 0인 경우, 가구수 컬럼으로 먼저 대체
master_df['세대수'] = np.where(master_df['세대수'] == 0, master_df['가구수'], master_df['세대수'])

# 위에서 계산한 결과가 여전히 0이면, 호수 컬럼으로 대체
master_df['세대수'] = np.where(master_df['세대수'] == 0, master_df['호수'], master_df['세대수'])

# 주차 대수 컬럼들의 합을 총주차수 컬럼에 저장
master_df['총주차수'] = master_df['옥내기계식대수'] + master_df['옥외기계식대수'] + master_df['옥내자주식대수'] + master_df['옥외자주식대수']

# 계산 끝난 불필요한 컬럼 제거
master_df.drop(['가구수', '호수', '옥내기계식대수', '옥외기계식대수', '옥내자주식대수', '옥외자주식대수'], axis=1, inplace=True)

# 데이터프레임 컬럼명 영문으로 변경
master_df.rename(columns=field_mapping, inplace=True)

# 전처리된 데이터프레임 확인
master_df
```
